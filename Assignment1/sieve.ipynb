```python
import time
import sys

def find_primes(n):
    
    try:
        # 1 GB for n = 10^9. A standard list would consume ~28GB.
        isPrime = bytearray([1]) * (n + 1)
    except MemoryError:
        print(f"Error: Failed to allocate {n+1} bytes (~{n/10**9:.1f} GB) of memory.")
        print("Please ensure you have enough free RAM.")
        return None, 0

    start_time = time.time()

    isPrime[0] = 0  
    isPrime[1] = 0  

    limit = int(n**0.5) + 1
    for p in range(2, limit):
        
        if isPrime[p]:
            for i in range(p * p, n + 1, p):
                isPrime[i] = 0

    largestPrime = 0
    for i in range(n - 1, 1, -1):  
        if isPrime[i]:
            largestPrime = i
            break
            
    end_time = time.time()
    time_spent = end_time - start_time
    
    return largestPrime, time_spent

if __name__ == "__main__":
    n = 1_000_000_000  

    print(f"Running Sieve for n = {n}...")
    
    largest_prime, duration = find_primes(n)
    
    if largest_prime is not None:
        print(f"Highest prime number below {n} is: {largest_prime}")
        print(f"Python execution time: {duration:.4f} seconds")
```

###
The total time taken for running the sieve to find the prime number in C is around 9 to 10 seconds on average meanwhile the time taken by python is  around 150 seconds which is way more than C.
